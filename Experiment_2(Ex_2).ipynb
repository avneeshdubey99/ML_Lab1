Correlation between engine-size and price

```
from scipy import stats
pearson_coef, p_value = stats.pearsonr(automod1['engine-size'], automod1['price'])
print("Pearson Correlation Coefficient =", pearson_coef, "\nP-value =", p_value) 
```

Since the p-value is < 0.001, the correlation between engine-size and price is statistically **significant**, and the **linear relationship** is very **strong**

We can see that the average price of cars with rear wheel drive is more than that of other cars.

Large proportion of the cars (38-39%) had rear wheel drive and the prices of these cars ranged from very low to very high, that is why we could be seeing a high average for cars with rear wheel drive.

We also see that the 4 wheel drive cars have a relatively high average even though there are very very few cars of that type (3-4%). We obtain this avergae because all the cars are relatively evenly priced and thus there is a lot less range in the price of the cars with 4 wheel drive which can affect the average price.

Making a pivot table 



```
temp = automod1[['drive-wheels','body-style','price']]
for_pivot = temp.groupby(['drive-wheels','body-style'], as_index=False).mean()
piv1 = for_pivot.pivot(index = 'drive-wheels', columns = 'body-style')
piv1
cols = automod1.describe(include=np.object).columns
empty = []
for i in cols:
    empty.append(i)
    for j in cols:
        if i==j or j in empty:
            continue
        else:
            temp = automod1.groupby([i,j])['price'].mean()
            print(temp, end='\n\n')
            temp = automod1[[i,j,'price']]
            for_pivot = temp.groupby([i,j], as_index=False).mean()
            piv = for_pivot.pivot(index = i, columns = j)
            piv.replace(np.nan, '-', inplace= True)
            print('Pivot Table')
            print(piv)
            print('*------*', end='\n\n')
```

One way anova test('drive-wheels' correlation 'price')

```
from scipy import stats
temp1 = automod1[['drive-wheels', 'body-style','price']]
temp = temp1[['drive-wheels', 'price']].groupby(['drive-wheels'])
temp
f_val, p_val = stats.f_oneway(temp.get_group('fwd')['price'], 
                              temp.get_group('4wd')['price'], 
                              temp.get_group('rwd')['price'])
print(f_val, p_val)
```
